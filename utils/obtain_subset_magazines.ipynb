```python
import pandas as pd
```

## Get subset of magazines from the east middle european region

* Languages: English, Polish, Hungarian, Czech, Slovenian
* Countries: Poland, Hungary, Czech Republic, Slovenia

```python
df = pd.read_csv('../data/source_datasets/mw_import_magazines.csv', encoding='utf-16')
```

```python
df.head()
```

```python
subset = df[df['Magazine[Origin]'].str.contains('Poland|Hungary|Czech Republic|Slovenia', na=False) | df['Magazine[Language]'].str.contains('Polish| Hungarian|Czech|Slovenian', na=False)]
subset = subset.sort_values(by=['Title'], key=lambda x: x.str.lower())
subset = subset.reset_index(drop=True)
```

```python
subset.head()
```

```python
subset.to_csv('../data/target_datasets/east_middle_europe_magazines.csv', index=False, encoding='utf-16')
```
